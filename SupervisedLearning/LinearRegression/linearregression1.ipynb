```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

## Data Preprocessing

```python
df = pd.read_csv("data.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.prices,color='red',marker='+')
```

## Model Train

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.prices)
```

```python
reg.predict([[3300]])
```

## Data Preprocessing

```python
d = pd.read_csv("areas.csv")
d.head(3)
```

```python
p = reg.predict(d)
d['prices'] = p
d
```

## Prediction Graph

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.prices,color='red',marker='*')
plt.plot(df.area,reg.predict(df[['area']]),color = 'green')
```
