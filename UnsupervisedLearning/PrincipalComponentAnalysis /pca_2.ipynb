## Data Preprocessing

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
df.shape
```

```python
df1 = df[df.Cholesterol<=(df.Cholesterol.mean()+3*df.Cholesterol.std())]
df1.shape
```

```python
df2 = df1[df1.Oldpeak<=(df1.Oldpeak.mean()+3*df1.Oldpeak.std())]
df2.shape
```

```python
df3 = df2[df2.RestingBP<=(df2.RestingBP.mean()+3*df2.RestingBP.std())]
df3.shape
```

```python
df3.ChestPainType.unique()
```

```python
df3.RestingECG.unique()
```

```python
df3.ExerciseAngina.unique()
```

```python
df3.ST_Slope.unique()
```

```python
df4 = df3.copy()
df4.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True)

df4.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)

df4.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)

df4.head()
```

```python
df5 = pd.get_dummies(df4, drop_first=True)
df5.head()
```

```python
df.head()
```

```python
X = df5.drop("HeartDisease",axis='columns')
y = df5.HeartDisease
X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
y
```

## model Train

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, train_size= 0.8)

model = RandomForestClassifier(n_estimators=20)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

## PCA to reduce dimensions

```python
pca = PCA(n_components=0.95)  # Retain 95% of variance
X_pca = pca.fit_transform(X_scaled)
X_pca
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, train_size=0.8, random_state=42)
```

```python
model_pca = RandomForestClassifier(n_estimators=20)
model_pca.fit(X_train_pca, y_train)

model_pca.score(X_test_pca, y_test)
```
